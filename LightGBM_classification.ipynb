```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score,recall_score,precision_score
```

```python
df=pd.read_csv('creditcard.csv')
df.head(2)
```

```python
X=df[[c for c in df.columns if c!= 'Class']]
y=df['Class']

skf = StratifiedKFold(n_splits=5,shuffle=True,random_state=101)
i=1
for train_index,test_index in skf.split(X,y):
    x_train,x_val = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train,y_val = y[train_index],y[test_index]
    
    clf=lgb.LGBMClassifier(n_estimators=1000,
                       learning_rate=0.005,
                       scale_pos_weight=10,

                       )

    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_val)
    
    score_precision = np.round(precision_score(y_val,y_pred),3)
    score_recal = np.round(recall_score(y_val,y_pred),3)
    score_f1= np.round(f1_score(y_val,y_pred),3)
    
    print("For fold {} - Precision is {}, Recall is {}, F1 score is {}".format(i,score_precision,score_recal,score_f1))
    i+=1
    
```
